# Calculate R

```python
def calc_r(method, var1, var2):
    """Calculates Pearson or Spearman R"""
    if method=="p":
        r_coeff, p_val = pearsonr(var1, var2)
        print(f'[{var1.name} & {var2.name}] \n  Pearson R Correlation Coefficient: {round(r_coeff,3)} \n  p-value: {round(p_val,3)}')
    else:
        r_coeff, p_val = spearmanr(var1, var2)
        print(f'[{var1.name} & {var2.name}] \n  Spearman R Correlation Coefficient: {round(r_coeff,3)} \n  p-value: {round(p_val,3)}')    
    return([r_coeff, p_val])
```

# Distribution of Numerical Variables

```python
def plot_box_num(var_num, display_name):
    
    churn1 = d[d['churn']=='Attrited Customer']
    churn0 = d[d['churn']=='Existing Customer']    
    
    fig = go.Figure()
    
    tr_b1 = go.Box(x=churn1[var_num],
                        name='Attrited', marker_color='rgb(64, 60, 115)')
    tr_b0 = go.Box(x=churn0[var_num],
                        name='Existing', marker_color='rgb(68, 130, 155)')
    
    fig.add_trace(tr_b1)
    fig.add_trace(tr_b0)
    
    fig.update_traces(opacity=0.5)
    fig.update_layout(title_text=f'Distribution of {display_name} by Churn Status',
                      height=250, width=800,
                      showlegend=True)
    fig.show()
```

```python
plot_box_num('age', 'Age')
```

```python
def plot_boxnhist_num(var_num, display_name):
    
    churn1 = d[d['churn']=='Attrited Customer']
    churn0 = d[d['churn']=='Existing Customer']    
            
    tr_b1 = go.Box(x=churn1[var_num],
                   name='Attrited', 
                   marker_color='rgb(64, 60, 115)', 
                   showlegend=False)
    tr_b0 = go.Box(x=churn0[var_num],
                   name='Existing', 
                   marker_color='rgb(68, 130, 155)', 
                   showlegend=False)
    tr_h1 = go.Histogram(x=churn1[var_num],
                         name='Attrited', 
                         marker_color='rgb(64, 60, 115)', 
                         showlegend=True)
    tr_h0 = go.Histogram(x=churn0[var_num],
                         name='Existing', 
                         marker_color='rgb(68, 130, 155)', 
                         showlegend=True)                        
    
    fig = make_subplots(rows=2, cols=1, row_heights=[0.20, 0.80], 
                        shared_xaxes=True, vertical_spacing=0.02)
    fig.add_trace(tr_b1, 1, 1)
    fig.add_trace(tr_b0, 1, 1)
    fig.add_trace(tr_h1, 2, 1)
    fig.add_trace(tr_h0, 2, 1)
        
    fig.update_traces(opacity=0.6)
    fig.update_layout(width=900, height=600, barmode='overlay',
                      legend=dict(orientation="h",
                                  yanchor="bottom", y=1,
                                  xanchor="right", x=0.99),
                      title_text=f'Distribution of {display_name} by Churn Status')
    fig.show()
```

```python
plot_boxnhist_num('age', 'Age')
```

# Distribution of Categorical Variables

```python
def plot_distr_cat_perc1(var, display_name, display_order):
    
    d_bar_full = d.groupby("churn")[var].value_counts().reset_index(name='churn_ct')

    churn1 = d_bar_full[d_bar_full['churn']=='Attrited Customer']
    churn0 = d_bar_full[d_bar_full['churn']=='Existing Customer']

    tr0 = go.Bar(x=churn1[var],
                 y=churn1['churn_ct'],
                 name='Attrited Customers',
                 marker_color='rgba(50, 136, 189, 0.5)')
    tr1 = go.Bar(x=churn0[var],
                 y=churn0['churn_ct'],
                 name='Existing Customers',
                 marker_color='rgba(244, 109, 67, 0.5)')

    data = [tr0, tr1]
    layout = go.Layout(barmode = "group",
                       title=f'<b>Distribution of {display_name} by Churn Status<b>',
                       xaxis=dict(title=display_name,
                                 titlefont_size=14,
                                 tickfont_size=12,
                                 tickangle=0,
                                 categoryorder='array',
                                 categoryarray=display_order),
                       yaxis=dict(title='Count of Churn Status',
                                 titlefont_size=14,
                                 tickfont_size=12),
                       legend=dict(x=0, y=1.0, bgcolor='rgba(255, 255, 255, 0)',
                                  bordercolor='rgba(255, 255, 255, 0)'),
                       bargap=0.2,
                       bargroupgap=0.1,
                       width=800,
                       height=500)
    fig = go.Figure(data=data, layout=layout)
    fig.show()
```

```python
def plot_var_to_baseline(var_cat, var_cat_display_name, var_cat_display_order):

    
    d_bar = d.groupby(var_cat)['churn'].value_counts().reset_index(name='ct')
    churn_ct_tot = d_bar.groupby(var_cat)['ct'].transform('sum')
    d_bar['%'] = d_bar['ct']/churn_ct_tot

    churn1 = d_bar[d_bar['churn']=='Attrited Customer'].reset_index(drop=True)
    ratio = (churn1.groupby('churn').ct.sum()[0])/d_bar.ct.sum()

    tr = go.Bar(x=churn1[var_cat],
                 y=churn1['%']-ratio,
                 name='Attrited Customers',
                 marker_color='rgba(50, 136, 189, 0.5)')

    data = [tr]
    layout = go.Layout(barmode = "stack",
                      title=f'Variance to Baseline of Attrited Customers by {var_cat_display_name}',
                      xaxis=dict(title=var_cat_display_name,
                                 titlefont_size=14,
                                 tickfont_size=12,
                                 tickangle=30,
                                 categoryorder='array',
                                 categoryarray=var_cat_display_order),
                      yaxis=dict(title='Variance to Baseline',
                                 titlefont_size=14,
                                 tickfont_size=12,
                                tickformat="%"),
                      bargap=0.2,
                      bargroupgap=0.1,
                      width=600,
                      height=400)
    fig = go.Figure(data=data, layout=layout)
    fig.show()
```

```python
order_cat = ['Unknown', 'Uneducated', 'High School', 
              'College', 'Graduate', 'Post-Graduate', 'Doctorate']
```

```python
def plot_distr_cat_perc5(var_cat, var_cat_display_name, var_cat_display_order):
    
#     top_labels = ['Unknown', 'Uneducated', 'High School', 
#               'College', 'Graduate', 'Post-Graduate', 'Doctorate']
    
    palette_orig = px.colors.sequential.deep[::-1]
    palette_new = [palette_orig[0],palette_orig[2],palette_orig[4],
                   palette_orig[5],palette_orig[6],palette_orig[8],palette_orig[9]]
    
    d_bar = d.groupby('churn')[var_cat].value_counts().reset_index(name='ct')
    churn_ct_tot = d_bar.groupby('churn')['ct'].transform('sum')
    d_bar['%'] = d_bar['ct']/churn_ct_tot*100

    churn1 = d_bar[d_bar['churn']=='Attrited Customer']
    churn0 = d_bar[d_bar['churn']=='Existing Customer']
    ct = churn0['ct']
    
    tr0 = go.Bar(x=churn0['%'],
                 y=["Existing" for i in range(7)],
                 name='Existing',
                 showlegend=False,
                 orientation='h',
                 hovertext=churn0['ct'],
                 text=churn0[var_cat],
                 texttemplate = "%{text} %{x:.0f}% <br>(%{hovertext:,d})",
                 textfont=dict(size=12, color='white'),
                 textposition = "inside",
                 marker=dict(color=palette_new, line=dict(color='rgb(248, 248, 249)', width=1)))
    tr1 = go.Bar(x=churn1['%'],
                 y=["Attrited" for i in range(7)],
                 name='Attrited',
                 showlegend=False,
                 orientation='h',
                 hovertext=churn1['ct'],
                 text=churn1[var_cat],
                 texttemplate = "%{text} %{x:.0f}% <br>(%{hovertext:,d})",
                 textfont=dict(size=12, color='white'),
                 textposition = "inside",
                 marker=dict(color=palette_new, line=dict(color='rgb(248, 248, 249)', width=1)))

    data = [tr0, tr1]
    layout = go.Layout(barmode = 'stack',
                       title=f'<b>Distribution of {var_cat_display_name} by Churn Status<b>',
                       xaxis=dict(title=f'<b>{var_cat_display_name} (%)<b>',
                                  titlefont_size=16,
                                  categoryorder='array',
                                  categoryarray=var_cat_display_order,
                                  tickmode='array',
                                  ticktext=['{0:.0f}'.format(i) + "%" for i in range(0,101,10)],
                                  tickfont_size=14,
                                  tickangle=0),
                       yaxis=dict(title='<b>Customer Churn Status<b>',
                                  titlefont_size=16,
                                  tickfont_size=14,
                                  tickangle=-90),
                       margin=dict(l=120, r=10, t=140, b=80),
                       bargap=0.2,
                       bargroupgap=0.1,
                       width=1300,
                       height=500)
    
    fig = go.Figure(data=data, layout=layout)
    fig.show()
```
